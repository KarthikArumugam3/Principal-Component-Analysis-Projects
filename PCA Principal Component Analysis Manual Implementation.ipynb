As we can see there are 30 features in this dataset and it becomes really hard for analysis and visualizations.

So for PCA we'll be performing Dimension reduction. We are not taking a subset of features from the 30 features, but instead we are going to convert the 30 features and the 30 dimensional space to some other component space that we can easily visualize.

### Scaling the Data
So the first step would be scale our data.

We'll Standardize the data, i.e. Making the mean of all the data points as 0.

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
scaled_X = scaler.fit_transform(df)
```

```python
scaled_X.mean()
```

So the mean is zero. It's just that the inabilty of the computer to interpret a large no of zeros after the decimal point.

### Covariance Matrix

```python
covariance_matrix = np.cov(scaled_X,rowvar=False)
```

We set the row variance/rowvar = False because really we just care about the variance across the features themselves, not every single row.

Now as in theaory we need to do Linear Transformation.

Numpy comes with a Linear algebra library built in that allows us to calculate the eigenvectors and eigenvalues automatically.

```python
eigen_values, eigen_vectors = np.linalg.eig(covariance_matrix)
```

```python
eigen_values
```

```python
eigen_vectors
```

```python
eigen_vectors.shape
```

So it makes sense right now that it is at higher dimension.

So we have our eigenvalues and eigenvectors and the next step is to choose how many components we would like to take from this.

----

So recall that we do calculate the eigenvetors upto "N" dimensions where "N" is he no of features and we're going to select a subset after we sort them by eigenvalues.

----

So, so far in our PCA process we have gone from N feature space ------> to N Principal Componenet space.

### So far the conversion made is not that useful as we still have the 30 features. So now we have to decide how many Principal Components to choose.
1. So we are going to choose N=2 for Principal Compenents from a 30 feature space. 
2. And after we calculate 30 Principal components, we're going to choose the top 2 based off the eigenvalues.
3. So we're going to sort the eigenvectors based off their associated eigenvalues and then choose the top 2.

#### We could have chosen the top 3 or top 4 but the whole point is to be able to visualize it. So we choose 2 as would be much easier to visualize.

```python
num_components = 2
```

So this is where we're going to go really heavy into the numpy because of the shapes of eigenvalues and eigenvectors, but the main idea here is we're simply sorting everything based off the eigenvalues and then we're going to grab the eigenvalues and eigenvectors up to N=2.

##### We'll use np.argsort here
argsort returns the index that would have sorted the array

Ex:- np.argsort([2,1,3]) 

o/p:- [1,0,2] i.e., it gives the index in a sequeance saying that is the i/p array were to sorted in this index order it would be sorted.

```python
np.argsort(eigen_values)
```

```python
# Now we will reverse this as we want to pick out the largest/greatest 2 values.
np.argsort(eigen_values)[::-1]
```

```python
# Grabbing only the our required no. of components 
np.argsort(eigen_values)[::-1][:num_components]
```

```python
# Hence our sorted key
sorted_key = np.argsort(eigen_values)[::-1][:num_components]
```

Hence we sorted the order and then made it descending and grabbed the first two largest values's keys.

```python
# We're just redefining the values again with the largest 2 i.e. what we are looking for
eigen_values, eigen_vectors = eigen_values[sorted_key],eigen_vectors[:,sorted_key]
```

So we just went from 30 features up there to 2 features based of the the sorted eigenvalues.

Now we still need to project our original data onto that eigenvector.

How to do that?

Remember that the DOT product is essentially just a projection measurement from one vector to another.

```python
# So the final Principal components we''l end up plotting:-
# So we are just projecting th eoriginal data onto the eigen vectors
# Taking the scaled_X which is 30 features and projecting them onto eigen_vetors which are 2 PCs
principal_components = np.dot(scaled_X,eigen_vectors)
```

```python
principal_components
```

```python
principal_components.shape
```

```python
scaled_X.shape
```

So we have reduced each component from the original dataset and here in principal_components each row coressponds to the original scaled_X

```python
# We can now plot out the data
plt.figure(figsize=(8,6),dpi=100)
plt.scatter(principal_components[:,0],principal_components[:,1]);
```

So we've been able to reduce a dataset from 30 features down to 2 principal components and the data itself is still seperable.

### So what that actually means is you didn't really need all the detail of those 30 original features.

#### We were able to reduce that to just 2 principal components and still have enough separation or enough information to separate this out into the two classes.

### So what does that mean as far as real life benefits?
This means that if we then wanted to run some sort of classification algorithm that scaled based of the number of dimensions it was running on, we could run PCA first, reduce it from 30 features to 2 dimensions, and then run our classification here and we would probably get very similar performance.

### So there's a lot of real world benefits as well, and the other one to consider is just the fact that we're saving a lot of space.
That is we went from 30 times 579 rows in original dataset to 2 times the 579 rows in PCA.
